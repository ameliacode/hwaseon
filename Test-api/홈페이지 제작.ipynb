```python
start = time.time()

driver = webdriver.Chrome('./chromedriver_win32/chromedriver.exe')
try:
    driver.get(f'https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&query={item}')
    sleep(0.1)
    r3 = driver.find_element_by_class_name('related_srch').text
except NoSuchElementException:
    r3 = ''

try:
    driver.find_element_by_class_name('greenbox').click()
    sleep(0.1)
    r1 = driver.find_element_by_class_name('kwd_lst._kwd_list').text
except NoSuchElementException:
    r1 = ''
    
try:
    driver.get(f'https://search.shopping.naver.com/search/all?where=all&frm=NVSCTAB&query={item}')
    sleep(0.1)
    driver.find_element_by_class_name('gnbSearch_inner__1ptdl').click()
    r2 = driver.find_element_by_class_name('autoComplete_layer_auto__29wPX.autoComplete_active__3O-sR').text
except NoSuchElementException:
    r2 = ''

r1 = r1.replace('\n추가','')
r1 = r1.split('\n')
r2 = r2.replace('\n도움말기능끄기','')
r2 = r2.split('\n')
r3 = r3.split('\n')
print(time.time()-start)
```

#### 수정된 부분

```python
### 손서영님께 문의. 카운트해서 오른쪽에 구분값. 직접 설명하기

print('네이버자동완성')
print(sorted(r1))
print('쇼핑자동완성')
print(sorted(r2))
print('연관검색어')
print(sorted(r3))
```

```python
naver_auto = {}
shop_auto = {}
rel_key = {}

for item in r1:
    naver_auto[item] = "네이버자동완성"
for item in r2:
    shop_auto[item] = "쇼핑자동완성"
for item in r3:
    rel_key[item] = "연관검색어"

all_items = sorted(list(set(r1 + r2 + r3)))
all_items
```

```python
for item in all_items:
    print(item, end=" | ")
    print( naver_auto[item] if item in naver_auto.keys() else "", end=" ")
    print( shop_auto[item] if item in shop_auto.keys() else "", end=" ")
    print( rel_key[item] if item in rel_key.keys() else "")
```
