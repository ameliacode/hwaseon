## 네이버 쇼핑 순위

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import json
import pandas as pd
import time
import urllib
```

```python
driver = webdriver.Chrome('./chromedriver_win32/chromedriver.exe')
time.sleep(0.5)
```

```python
encText = urllib.parse.quote("폼클렌징") ## 검색하고자 하는 키워드
```

```python
naver_shop = 'https://search.shopping.naver.com/search/all?'+'frm=NVSCTAB'+'&origQuery='+encText+'&pagingIndex={}'+'&pagingSize=40&productSet=total'+'&query='+encText+'&sort=rel&timestamp=&viewType=list'
```

```python
urls = [naver_shop.format(i) for i in range(1,11)]
all_result = []
for _url in urls: 
    driver.get(_url)
    result = WebDriverWait(driver, 10).until(EC.presence_of_element_located(
        (By.XPATH, "//script[@type='application/json']"))).get_attribute('innerHTML')
    time.sleep(1)
    result_json = json.loads(result.strip())
    all_result.append(result_json)
```

```python
# 불필요한 데이터 제거
def preprocessing(result_json):

    delete_keys = ["assetPrefix","customServer","gssp","dynamicIds","runtimeConfig","isFallback","buildId", "query",
                  "page"]

    for key in delete_keys:
        try:
            result_json.pop(key)
        except:
            pass
        
    return result_json
```

```python
#json에서 뽑고자 하는 정보
result_json["props"]["pageProps"]["initialState"]["products"]["list"][0]["item"]["rank"] #순위
result_json["props"]["pageProps"]["initialState"]["products"]["list"][0]["item"]["productTitle"] #상품명
result_json["props"]["pageProps"]["initialState"]["products"]["list"][0]["item"]["shopNNo"] #NVID
```

```python
def processing_result(result_json): #결과 Json array화
    result_list = result_json["props"]["pageProps"]["initialState"]["products"]["list"]
    total_list = []
    for item in result_list:
        temp = {}
        title = item["item"]["productTitle"]
        rank = item["item"]["rank"]
        bundle = item["item"]["shopNNo"]
        store = item["item"]["mallName"]

        try:
            ad = item["item"]["adId"] 
        except:
            ad = None
        try:
            temp["id"] = item["item"]["id"]
        except:
            temp["id"] = None

        temp["title"] = title
        temp["rank"] = rank
        temp["bundle"] = bundle
        temp["store"] = store
        temp["ad"] = ad

        temp["mall_list"] = None

        try:
            mall_list = []
            lowMallList = item["item"]["lowMallList"]
            for i in lowMallList:
                mall_temp = {}
                mall_temp[i["nvMid"]] = i["name"]
                mall_list.append(mall_temp)
        except:
            #print(json.dumps(item["item"],indent = 4, ensure_ascii=False))
            pass

        temp["mall_list"] = mall_list

        total_list.append(temp)
    return total_list
```

```python
final_result = []
for each_json in all_result:
    each_json = preprocessing(each_json)
    final_result.append(processing_result(each_json))
```

```python
final_result = sum(final_result, [])
final_result
```

## 예상판매량

```python
def get_category(item):
    categories = item.find_elements_by_tag_name("a")
    return [i.text for i in categories]
```

```python
def get_link(item):
    link = item.find_element_by_tag_name("a").get_attribute("href")
    return link
```

```python
def get_title(item):
    title = item.find_element_by_tag_name("a").text
    return title
```

```python
def get_shipping(item):
    shipping_fee = 0
    fee = item.find_element_by_tag_name("em").text
    if not fee[4:].startswith("무료"):
        shipping_fee = fee[4:].replace("원","")
        shipping_fee = shipping_fee.replace(",","")
    return int(shipping_fee)
```

```python
def get_store_name(item):
    store = item.find_element_by_class_name("basicList_mall__sbVax").text
    return store
```

```python
def get_total_price(driver):
    get_buy_nums = driver.find_elements_by_class_name("Zye_YMu5OB")
    get_price = int(driver.find_element_by_class_name("_1LY7DqCnwR").text.replace(",",""))
    total = sum([int(i.text.split("건")[0]) for i in get_buy_nums])
    total_price = get_price * total
    return total_price
```

```python
def get_sales(driver):
    get_buy_nums = driver.find_elements_by_class_name("Zye_YMu5OB")
    total = sum([int(i.text.split("건")[0]) for i in get_buy_nums])
    return total
```

```python
def get_price(item):
    get_price = item.find_element_by_tag_name("span").text
    get_price = get_price.replace("원","")
    get_price = int(get_price.replace(",",""))
    return get_price
```

```python
n_shop_url = "https://search.shopping.naver.com/search/all?frm=NVSHCHK&origQuery={}&pagingIndex=1&pagingSize=40&productSet=checkout&query={}&sort=rel&timestamp=&viewType=list".format(encText, encText)
driver = webdriver.Chrome("./chromedriver_win32/chromedriver.exe")
driver.get(n_shop_url)
driver.execute_script("window.scrollTo(0,document.body.scrollHeight);") #끝까지 스크롤다운
time.sleep(2)
```

```python
get_item_list = driver.find_element_by_class_name("list_basis")
items = get_item_list.find_elements_by_tag_name("div")
all_items_info = {}
info = {}
for item in items:
    class_name = item.get_attribute("class")
    if class_name.startswith("basicList_title"):
        title = get_title(item)
        info["link"] = get_link(item)
    elif class_name.startswith("basicList_mall_title"):
        info["store"] = get_store_name(item)
    elif class_name.startswith("basicList_depth"):
        info["category"] = get_category(item)
    elif class_name.startswith("basicList_mall_area"):
        info["shipping"] = get_shipping(item)
    elif class_name.startswith("basicList_price_area"):
        info["price"] = get_price(item)
    elif class_name.startswith("basicList_mall_grade"):
        info["price_shipping"] = info["shipping"] + info["price"]
        all_items_info[title]= info
        info = {}
```

```python
for key,values in all_items_info.items():
    print(key)
    for title, value in values.items():
        print(title, value)
    print("------------------------")
print(len(all_items_info)) #통상적으로 40이상이 나와야함(이 부분이 안정적이지가 않음)
```

```python
# 위에서 1페이지에서 파싱한 링크 각각 판매량 조사
for title,info in all_items_info.items():
    begin = time.time()
    driver.get(info["link"])
    time.sleep(1)
    try:      
        sales = get_sales(driver)
    except:
        sales = "Unknown"
        print("Time elapsed: {:.3f}s".format(time.time() - begin))
    info["sales"] = sales
    all_items_info[title] = info
```
